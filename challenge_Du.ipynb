```python
##classofieur:kmeans par scripts:on choisi les premiere deux channel 
mean1=np.array(pca_12[0,:])
                   
mean2=np.array(pca_12[2,:])
                    
mean3=np.array(pca_12[5,:])

mean4=np.array(pca_12[8,:])

mean5=np.array(pca_12[14,:])

mean6=np.array(pca_12[4,:])
```

```python
nm1=0
nm2=0
nm3=0
nm4=0
nm5=0
nm6=0
for i in range (1000):
    test = np.array(pca_12[i,:])
    err = [np.linalg.norm(test-mean1),np.linalg.norm(test-mean2),np.linalg.norm(test-mean3),np.linalg.norm(test-mean4),np.linalg.norm(test-mean5),np.linalg.norm(test-mean6)]
    if err[0] == min(err):
        nm1 += 1
        mean1 = (mean1*nm1+test)/(nm1+1)
    elif err[1] == min(err):
        nm2 += 1
        mean2 = (mean2*nm2+test)/(nm2+1)
    elif err[2] == min(err):
        nm3 += 1
        mean3 = (mean3*nm3+test)/(nm3+1)
    elif err[3] == min(err):
        nm4 += 1
        mean4 = (mean4*nm4+test)/(nm4+1)
    elif err[4] == min(err):
        nm5 += 1
        mean5 = (mean5*nm5+test)/(nm5+1)
    elif err[5] == min(err):
        nm6 += 1
        mean6 = (mean6*nm6+test)/(nm6+1)
        
```

```python
answer=[1 for i in range (1000)]
for i in range (1000):
    test = np.array(pca_12[i,:])
    err = [np.linalg.norm(test-mean1),np.linalg.norm(test-mean2),np.linalg.norm(test-mean3),np.linalg.norm(test-mean4),np.linalg.norm(test-mean5),np.linalg.norm(test-mean6)]
    if err[5] == min(err) or err[1]==min(err):
        answer[i]= 0
```
